```python
import pandas as pd
import folium
```

*Dicionário das informações do dataset*

```python
data = pd.read_csv("data/Data_Dictionary.csv", sep = ";")
data = data[['FIELD', 'DESCRIPTION']].dropna() # drop rows with NaN values
print(data)
```

*Importando todas cidades do dataset*

```python
full_br_cities = pd.read_csv("data/BRAZIL_CITIES.csv", sep = ";")
```

*Agrupando cidades por estado*

```python
br_cities = full_br_cities[['STATE', 'CITY', 'LAT', 'LONG', 'POST_OFFICES']]
br_cities = br_cities.groupby('STATE')

print(br_cities.head())
```

*Buscando cidades de MG e Leopoldina*

```python
print(br_cities.get_group('MG'))
print(br_cities.get_group('MG')[br_cities.get_group('MG')['CITY'] == 'Leopoldina'])
```

*População das cidades*

```python
br_pop_cities = full_br_cities[['STATE', 'CITY', 'IBGE_RES_POP_BRAS']]
br_pop_cities = br_pop_cities.groupby('STATE')

print(br_pop_cities.get_group('MG'))
print(br_pop_cities.get_group('MG')[br_pop_cities.get_group('MG')['CITY'] == 'Leopoldina'])
```

*Criando mapa com markers nas cidades de Minas Gerais (MG)*

```python
mg_cities = br_cities.get_group('MG')
map = folium.Map(location=[-14.1558097,-49.1357842], zoom_start=4) 
for index, row in mg_cities.iterrows():
    folium.Marker([row['LAT'], row['LONG']], popup=row['CITY']).add_to(map)

map
```
